Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Product':['Laptop', 'Mac', 'Printer','Mac', 'Printer'],
                   'Region':['North', 'South', 'East', 'West', 'North'],
                   'Sales':[300, 400, 500, 600, 700]})
```

```python
group_by = df.groupby('Product')['Sales'].sum()
```

```python
group_by
```

```python
plt.pie(group_by, labels = group_by.index, autopct = '%1.1f%%')
```
