```python
import pandas as pd
```

```python
data = {
    'Student':['Arun', 'Arun', 'Meena', 'Meena'],
    'Marks': [70, 80, 75, 85]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df
```

TOTAL COUNT

```python
total_count = df['Student'].count()
```

```python
print(total_count)
```

MEAN

```python
mean_marks = df['Marks'].mean()
```

```python
print(mean_marks)
```

```python
median_marks = df['Marks'].median()
```

```python
print(median_marks)
```

GROUP BY

```python
total_groups = df.groupby('Student')['Marks'].sum()
```

```python
print(total_groups)
```
