```python
import squarify
import matplotlib.pyplot as plt
```

```python
sizes = [40, 30, 20, 10]
labels = ['IT', 'HR', 'Sales', 'Finance'] # Categorical
```

NOTE: We can also specify which colors to use

```python
squarify.plot(sizes = sizes, label = labels, alpha = 0.5)
plt.title('Distribution of Employees in a Company')
plt.show()
```
